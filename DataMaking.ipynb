# Make the Data
## Parking lots
### 1000*1200 
* Amount
* (Start_time, End_time)

```python
Park_n = 1200
Park_time = []
for i in range(Park_n):
    s = np.random.randint(5,6) + np.random.rand()
    e = np.random.randint(23,24) + np.random.rand()
    Park_time.append([s,e])
```

## Car users
* (Start_time, End_time)
* Preference list:P_list
* User_cost:Car_C

```python
Car_n = 1000
Car_Time = []
for i in range(Car_n):
    temp = np.random.randint(0,3)
    if temp == 0:
        s = np.random.rand() + np.random.randint(6,9)#[6,9)
        e = np.random.rand() + np.random.randint(11,14)
        Car_Time.append([s,e])
    elif temp == 1:
        s = np.random.rand() + np.random.randint(12,14)
        e = np.random.rand() + np.random.randint(18,23)
        Car_Time.append([s,e])
    else:
        s = np.random.rand() + np.random.randint(6,20)
        e = np.random.rand() + np.random.randint(int(s)+1,23)
        Car_Time.append([s,e])
    ##make the time of the cars
```

```python
Car_C = []
for i in range(Car_n):
    temp1 = []
    for j in range(Park_n):
        temp1.append(np.random.rand())
    Car_C.append(temp1)
```

```python
P_list = []
for i in range(len(Car_C)):
    a = Car_C[i]
    b = sorted(enumerate(a),key = lambda x:x[1])
    P_list.append(b)
```

```python
Car_Time = sorted(Car_Time)
```

## Save date

```python
conut = '5_1200'
```

```python
# Time
file_path = 'Data'
name = ['start_time','end_time']
test = pd.DataFrame(columns = name,data = Car_Time)
test1= pd.DataFrame(columns = name,data = Park_time)
test.to_csv(file_path+'/'+str(conut)+'Car_time.csv',index=False)
test1.to_csv(file_path+'/'+str(conut)+'Park_time.csv',index=False)
```

```python
# User cost
name = []
for i in range(Park_n):
    temp = 'Park_'+str(i)
    name.append(temp)
test3 = pd.DataFrame(columns = name, data = Car_C)
test3.to_csv(file_path+'/'+str(conut)+'User_cost.csv',index=False)
```

```python
# Preference list
name = []
for i in range(Park_n):
    temp = 'Park_'+str(i)+'_Pre'
    name.append(temp)
test3 = pd.DataFrame(columns = name, data = P_list)
test3.to_csv(file_path+'/'+str(conut)+'User_PreferenceList.csv',index=False)
```
