```python
Car_n = 1000
Park_n = 1200
Match = []
for i in range(Car_n):
    Match.append([row_index[i],col_index[i]])
```

```python
Park_sumtime = 0
Car_sumtime = 0
for i in range(Car_n):
    temp1 = Car_time.loc[i][1] - Car_time.loc[i][0]
    Car_sumtime = Car_sumtime + temp1
    j = Match[i][1]  
    temp2 = Park_time.loc[j][1] - Park_time.loc[j][0]
    Park_sumtime = Park_sumtime + temp2
```

```python
Pre_ID = []
for i in range(Car_n):
    temp = []
    for j in range(Park_n):
        k = int(User_PreList.loc[i][j].split(',')[0][1:])
        temp.append(k)
    Pre_ID.append(temp)
```
